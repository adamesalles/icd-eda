# Principais Franquias da Nintendo

Nessa parte do trabalho teremos como objetivos analisar 3 das principais franquias da empresa Nintendo, tanto individualmente quanto entre elas. 

As três franquias escolhidas foram:
 - The Legend of Zelda
 - Mario Bros.
 - Pokémon
 
Com isso, iniciaremos nossas análises.

```python
import pandas as pd #pip install --upgrade pandas
import numpy as np #pip install --upgrade numpy
import matplotlib.pyplot as plt #pip install --upgrade matplotlib
from db import *
import seaborn as sns #pip install --upgrade seaborn
```

Criando um dataframe com os dados sem alteração

```python
vgsales = get_db()
vgsales
```

## Criação dos dataframes

Para essa análise, consideramos somente os jogos que têm uma palavra chave em seu título que caracterizaria esse jogo como um jogo principal (ou, ao menos, importante) de sua respectiva franquia. Também poderemos ver quais são os jogos mais vendidos de cada franquia, após a criação de seu respectivo dataframe.

### The Legend of Zelda

Para essa franquia, selecionamos somente as que contém o nome `Zelda` em seu título.

```python
zelda = vgsales[vgsales.Name.str.contains("Zelda")] # Filtra os dados para apenas os que contêm a string Zelda. Utiliza o fato de que strings são listas de caracteres em python.
zelda.head() # Top 5 jogos mais vendidos
```

### Mario Bros.

O caso do Mario Bros. é o mais notável, pois há algun jogos do universo Mario que não contêm seu nome no título. Para exemplificar, existe a sequência "Wario Land". Wario é um vilão da franquia, e ganhou seu próprio jogo. O mesmo acontece com Luigi e outros pesonagens.

Contudo, para seguir o critério, selecionaremos somente os jogos que contêm o nome `Mario`. Foi necessário garantir que a palavra não estava contida em outra, neste caso, colocamos a condição que a palavra `Marionette` não poderia estar inclusa.

```python
mario = vgsales[(vgsales.Name.str.contains("Mario")) & ~(vgsales.Name.str.contains("Marionette"))] # Igual ao anterior, mas também adicionando a condição de que não pode conter Marionette
mario.head() # Top 5 jogos mais vendidos
```

### Pokémon

Para os jogos de Pokémon, houve a necessidade de escolher duas palavras-chave por uma inconsistência de grafia. Nesse banco de dados, tanto `Pokémon` quanto `Pokemon` estavam presentes entre os jogos. Portanto, decidimos escolher ambas.

```python
pokemon = vgsales[vgsales.Name.str.contains("Pokemon") | vgsales.Name.str.contains("Pokémon") ] # Operador 'or'
pokemon.head()  # Top 5 jogos mais vendidos
```

### Unificado

Para as comparações entre as franquias, acreditamos que será importante criar um único dataframe unindo todos os jogos. Para poder diferenciá-los, criaremos a coluna `Franchise` com sua respectiva franquia.

```python
# Criando uma coluna com o valor de cada franquia em cada dataframe
zelda = zelda.assign(Franchise = "The Legend of Zelda")
mario = mario.assign(Franchise = "Mario Bros.")
pokemon = pokemon.assign(Franchise = "Pokémon")

# Criando o dataframe franchises e adicionando a ele cada um dos outros df.
franchises = zelda.append(mario)
franchises = franchises.append(pokemon)
franchises = franchises.sort_values(by="Rank",ascending=True)

franchises.head() # Top 5 jogos mais vendidos
```

## Análises

### Vendas

Num primeiro momento, acreditamos que é interessante analisar a venda ao longo dos anos. Portanto, vamos fazer um _linechart_ para cada franquia e, posteriormente, uni-los. 


```python
zelda_sy = zelda.groupby(["Year"],as_index=False)[["Global_Sales"]].sum() # Agrupando por ano e fazendo a soma das vendas globais.
zelda_sy = zelda_sy[zelda_sy.Year != -1] # Conferindo se não há anos -1 (-1 significa que não há a informação de ano)
zelda_sy.head()
```

The Legend of Zelda, em comparação a si mesmo, tem altos e baixos ao longo dos anos. O ponto notável de sua trajetória é em 1998, quando "The Legend of Zelda: Ocarina of Time" foi lançado. Esse jogo mantém a maior pontuação no site de análises <https://www.metacritic.com/> desde quando fora analisado.

```python
sns.set_theme(style="darkgrid") # Tema do seaborn (se não funcionar, lembrar de atualizar o módulo)
fig, ax = plt.subplots(figsize=(12,8)) # Cria a figura e eixo
sns.lineplot(ax = ax, x="Year", y="Global_Sales", color="green", data=zelda_sy) # Lineplot e seus parâmetros
plt.xticks(range(1986,2015,2)) # Faz com que a divisão os anos seja discreta, também serve como padronização.
plt.title("Global Sales of The Legend of Zelda Games Through the Years") # Título
ax.set(ylabel='Global Sales ($10^6$)') # Rótulo eixo y
plt.show()
```

```python
# O mesmo processo será feito em Mário e Pokémon.
mario_sy = mario.groupby(["Year"],as_index=False)[["Global_Sales"]].sum()
mario_sy = mario_sy[mario_sy.Year != -1]
mario_sy.head()
```

Mario Bros. tem números expressivos basicamente ao longo de todos os anos. Ainda mais se olharmos a escala, que mostra os mínimos ainda na casa dos 5/6 milhões. Há dois ápices da franquia. Um em 1985, com o "Super Mario Bros.", que foi uma revolução nos jogos. Além do período entre 2006 e 2008, que foi quando Super Mario Galaxy, New Super Mario Bros. (Wii) e Mario Kart Wii foram lançados, dois títulos que acompanharam o sucesso do console Nintendo Wii (o mais vendido de sua geração). 

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(ax = ax, x="Year", y="Global_Sales", color="red", data=mario_sy)
plt.xticks(range(1982,2015,2))
plt.title("Global Sales of Mario Bros. Games Through the Years")
ax.set(ylabel='Global Sales ($10^6$)')
plt.show()
```

```python
pokemon_sy = pokemon.groupby(["Year"],as_index=False)[["Global_Sales"]].sum()
pokemon_sy.head()
```

A franquia de Pokémon, que é a mais nova dentre as três, também manteve relativo sucesso, embora nunca tenha voltado à glória do fim dos anos 90, com Pokémon Red/Blue e Gold/Silver. 

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(ax = ax, x="Year", y="Global_Sales", color="orange", data=pokemon_sy)
plt.xticks(range(1996,2015,2))
plt.title("Global Sales of Pokémon Games Through the Years")
ax.set(ylabel='Global Sales')
plt.show()
```

```python
franchises_sy = franchises.groupby(["Year", "Franchise"],as_index=False)[["Global_Sales"]].sum()
franchises_sy = franchises_sy[franchises_sy.Year != -1]
franchises_sy.head()
```

Agora, analisando todos em mesma escala e em mesma figura, podemos ver que Mario Bros. e, em poucos momentos, Pokémon lideram as vendas. Definitivamente, embora tenha jogos muito bem avaliados, The Legend of Zelda não ultrapassa nem a marca dos 10 milhões de vendas. Enquantos os outros costumam estar sempre acima dela.

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(ax = ax, x="Year", y="Global_Sales", hue='Franchise',palette=['r','g','orange'], data=franchises_sy)
plt.xticks(range(1982,2015,2))
plt.title("Global Sales of Nintendo Main Franchises Through the Years")
ax.set(ylabel='Global Sales')
plt.show()
```

### Vendas por Região

Nesse gráfico podemos comparar como as franquias se saem em cada região. Nesse gráfico desconstrói algumas ideias preliminares que costumamos ter. Por exemplo, só há um grande domínimo de Mario Bros. nos Estados Unidos. Nos outros mercados, principalmente no Japão, não há uma diferença tão grande entre Mario Bros. e Pokémon. 

Novamente, ao falar de The Legend of Zelda, vemos que não há melhora nem em seu mercado natal, o Japão.

```python
franchises_mrkt = franchises.groupby(["Franchise"],as_index=False)[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()

fig, ax = plt.subplots(2, 2,figsize=(12,8), sharey = True)
sns.barplot(ax = ax[0, 0], x="Franchise", y="NA_Sales", data=franchises_mrkt,ci=None,palette=['r','orange','green'])
sns.barplot(ax = ax[0, 1], x="Franchise", y="EU_Sales", data=franchises_mrkt,ci=None,palette=['r','orange','green'])
sns.barplot(ax = ax[1, 0], x="Franchise", y="JP_Sales", data=franchises_mrkt,ci=None,palette=['r','orange','green'])
sns.barplot(ax = ax[1, 1], x="Franchise", y="Other_Sales", data=franchises_mrkt,ci=None,palette=['r','orange','green'])
plt.suptitle('Sales of Nintendo Main Franchises divided by Region')
plt.show()
```

### Média de vendas por jogo

Contudo, é imporante avaliar que não há a mesma quantiade de jogos entre as franquias, a partir de uma visualização textual, é possível comparar a média de vendas por jogo.

Dessa forma, vemos que, embora haja uma diferença enorme entre a venda total das franquias, e que Mario Bros. as lidere com sobra. Em média, Pokémon se sai melhor e The Legend of Zelda não está tão atrás como estava em valores absolutos.

```python
franchises_as = franchises.groupby(["Franchise"],as_index=False)[["Name"]].count()
franchises_sales = franchises.groupby(["Franchise"],as_index=False)[["Global_Sales"]].sum()
franchises_as = franchises_as.assign( Global_Sales = franchises_sales["Global_Sales"])
franchises_as = franchises_as.assign( Sales_Average = lambda df: df["Global_Sales"]/df["Name"])
franchises_as
```

### Jogos por Gênero

Outro modo de avaliar os jogos, é em relação a seus gêneros. No caso, a partir de gráficos de barra nós conseguiremos visualizar a forma em que eles são distribuídos.

```python
#contando a quantidade de jogos
franchises_qg = franchises.groupby(["Franchise","Genre"],as_index=False)[["Name"]].count()
franchises_qg.head()
```

Podemos ver que em jogos de plataforma (como mostrado anteriormente, o gênero mais vendido por muitos anos), por exemplo, só há jogos do Mario Bros., enquanto em Role-Playing, Pokémon domina.

The Legend of Zelda se concentra somente em Ação e Aventura.

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(ax = ax, x="Genre", y="Name", data=franchises_qg, hue="Franchise", palette=['r','orange','green'])
ax.set(ylabel='Games', title='Games of Nintendo Main Franchises per Genre')
plt.show()
```

### Jogos por Plataforma

Novamente, jogos carregam consigo, além de seu Gênero, as plataformas onde estão disponíveis. Ao longo da história dos games, muitas empresas surgiram ou deixaram de existir no cenário de plataformas. Por exemplo, a SEGA deixou de fazer consoles após um fracasso comercial no começo dos anos 2000 (com o Dreamcast), enquanto a Microsoft surgiu na mesma época com uma proposta que ganhou o mercado. No caso da Nintendo, houve altos e baixos, mas sempre teve sucesso em seus consoles portáteis. 

```python
franchises_gp = franchises.groupby(["Franchise","Platform"],as_index=False)[["Name"]].count()
franchises_gp.head()
```


Em todos os consoles, com exceção do Nintendo DS, o Mario Bros. possui mais títulos. No Nintendo DS, o Pokémon lidera. No gráfico posterior a esse, entenderemos um fator decisivo.


```python
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(ax = ax, x="Platform", y="Name", data=franchises_gp, hue="Franchise", palette=['r','orange','green'])
ax.set(ylabel='Games', title='Games of Nintendo Main Franchises per Genre')
plt.show()
```

Agora, como dito anteriormente, podemos dividir os jogos entre os de plataformas portáteis e plataformas de mesa. No caso, esse barplot mostra que a relação entre os dois tipos de jogos em Mario Bros. e em The Legend of Zelda é semelhante. Mas, em Pokémon, há uma grande diferença. Isso reflete no gráfico anterior. Principalmente devido ao fato de Pokémon ser uma franquia que surgiu nos portáteis, o que levou à tradição de só lançarem jogos principais (naturalmente, os que vendem mais) nesse tipo de console.

```python
franchises_ptb = franchises_gp.assign(
    Kind = lambda dataframe: dataframe['Platform'].map(lambda platform: "Handheld Console" if platform in ('3DS','GB','GBA','DS') else "Home Console") 
)

franchises_ptb = franchises_ptb.groupby(["Franchise","Kind"],as_index=False)[["Name"]].sum()

fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(ax = ax, x="Franchise", y="Name", data=franchises_ptb, hue="Kind",ci=None)
ax.set(ylabel='Games', title='Games of Nintendo Main Franchises divided by the kind of its console')
plt.show()
```
